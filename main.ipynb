```python
product_list = ['ZS','ZM','ZW','ZC','CL','ZL']

for product in product_list:
    product_parameter = product + '=F'
    ret = data_loader(contract_name=product_parameter,period = '365d')
    df_daily = ret['df_daily']
    df_60 = ret['df_60']
    df_240 = ret['df_240']

    df_daily,df_240,df_60 = final_signal_generator(df_daily,df_240,df_60)
    trades_df,equity_curve,drawdown = trades_generator(df_60)
    metrics = analyze_trades(trades_df,equity_curve,drawdown)
    metrics['product'] = product
    # print(metrics)
    for key, value in metrics.items():
        print(f"{key}: {value}")
    
    list_dict = [
        {   
            'name' : 'equity_curve',
            'series' : equity_curve,
            'color' : 'blue',
        },
        {   
            'name' : 'drawdown_curve',
            'series' : drawdown,
            'color' : 'red',
        }
    ]
    create_line_chart(df_60,list_dict)
```
